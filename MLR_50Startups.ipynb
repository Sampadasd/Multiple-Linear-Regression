```python
#Collinearity Problem Check

rsq_r=smf.ols("RDS~Administration+MKTS",data=startups_rename).fit().rsquared
vif_r=1/(1-rsq_r)

rsq_a=smf.ols("Administration~RDS+MKTS",data=startups_rename).fit().rsquared
vif_a=1/(1-rsq_a)

rsq_m=smf.ols("MKTS~RDS+Administration",data=startups_rename).fit().rsquared
vif_m=1/(1-rsq_m)

# Putting the values in Dataframe format
d1={'Variables':['RDS','Administration','MKTS'],'Vif':[vif_r,vif_a,vif_m]}
Vif_df=pd.DataFrame(d1)
Vif_df
```

None variable has VIF>20, No Collinearity, so consider all varaibles in Regression equation

```python
#Residual Analysis
sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

```python
# Test for Homoscedasticity or Heteroscedasticity
def standard_values(vals) : return (vals-vals.mean())/vals.std()  #z = (x - mu)/sigma
```

```python
plt.scatter(standard_values(model.fittedvalues),standard_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('Standardized fitted values')
plt.ylabel('Standardized residual values')
plt.show() 
```

```python
# Test for errors or Residuals Vs Regressors
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'RDS',fig=fig)
plt.show()
```
